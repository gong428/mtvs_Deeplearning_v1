```python
import os
import random
from PIL import Image, ImageEnhance

def adjust_saturation_brightness(image_path, output_path):
    """
    주어진 이미지의 채도와 밝기를 랜덤으로 조정하고 저장합니다.

    Args:
    image_path (str): 원본 이미지 경로
    output_path (str): 수정된 이미지를 저장할 경로
    """
    # 이미지 열기
    img = Image.open(image_path)
    
    # 랜덤한 채도와 밝기 값을 생성 (0.5 ~ 1.5 범위)
    saturation_factor = random.uniform(0.5, 1.5)
    brightness_factor = random.uniform(0.5, 1.5)
    
    # 채도 조정
    enhancer = ImageEnhance.Color(img)
    img = enhancer.enhance(saturation_factor)
    
    # 밝기 조정
    enhancer = ImageEnhance.Brightness(img)
    img = enhancer.enhance(brightness_factor)
    
    # 이미지 저장
    img.save(output_path)

def process_images_in_folder(input_folder, output_folder):
    """
    입력 폴더의 모든 이미지에 대해 채도와 밝기를 랜덤으로 조정합니다.

    Args:
    input_folder (str): 원본 이미지가 저장된 폴더
    output_folder (str): 수정된 이미지를 저장할 폴더
    """
    # 출력 폴더가 없으면 생성
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # 폴더 내 모든 파일에 대해 처리
    for filename in os.listdir(input_folder):
        if filename.endswith((".jpg", ".jpeg", ".png")):
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)
            adjust_saturation_brightness(input_path, output_path)
            print(f"Processed {filename}")

# 사용 예시
input_folder = "../cut_datas/1003499.jpg"
output_folder = "../random_image/1003499.jpg"
process_images_in_folder(input_folder, output_folder)
```

```python
import os
import random
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt

def random_adjust_image(input_path):
    """
    이미지를 랜덤하게 채도와 밝기를 조정하여 출력합니다.
    
    :param input_path: 입력 이미지 파일 경로
    """
    try:
        # 이미지 열기
        image = Image.open(input_path)
    except FileNotFoundError:
        print(f"Error: 파일 {input_path}이(가) 존재하지 않습니다.")
        return
    except IOError:
        print(f"Error: 파일 {input_path}을(를) 열 수 없습니다. 지원되지 않는 형식이거나 손상된 파일입니다.")
        return

    # 랜덤하게 채도 조정 (0.7 ~ 1.3)
    saturation_factor = random.uniform(0.7, 1.3)
    enhancer = ImageEnhance.Color(image)
    image = enhancer.enhance(saturation_factor)

    # 랜덤하게 밝기 조정 (0.7 ~ 1.3)
    brightness_factor = random.uniform(0.7, 1.3)
    enhancer = ImageEnhance.Brightness(image)
    image = enhancer.enhance(brightness_factor)

    # 이미지 출력
    plt.imshow(image)
    plt.axis('off')  # 축을 표시하지 않음
    plt.show()

if __name__ == "__main__":
    input_image_path = "../cut_datas/1003499.jpg"
    random_adjust_image(input_image_path)
```
