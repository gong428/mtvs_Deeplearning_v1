```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
import os
import requests
import base64
# 검색어를 입력받는 함수
def get_search_query():
    return input("검색어를 입력하세요: ")

# 구글 검색을 수행하는 함수
def google_search(query):
    # Chrome 드라이버 설정
    service = Service(ChromeDriverManager().install())
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    
    # 웹 드라이버 초기화
    driver = webdriver.Chrome(service=service, options=options)

    try:
        # 구글 홈페이지로 이동
        driver.get("https://www.google.com")

        # 검색창에 검색어 입력
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys(query)
        search_box.submit()

        # 검색 결과 페이지가 로드될 때까지 기다림
        time.sleep(2)  # 검색 결과 페이지가 로드될 시간을 주기 위해 일시적으로 sleep 사용
        
        print(f"'{query}'에 대한 구글 검색이 완료되었습니다.")

    except Exception as e:
        print(f"Error during search: {e}")
    '''finally:
        # 드라이버 종료
        driver.quit()'''
    
    return driver
#링크 클릭
def click_button(element,driver):
    try:
        driver.execute_script("arguments[0].click();", element)
        time.sleep(2)
    except Exception as e:
        print(f"Error occurred: {e}")   
# 탭 클릭
def cilck_tab(num,driver):# 0:전체, 1:이미지, 2: 쇼핑, 3: 동영상, 4:뉴스, 5:지도, 6:도서
    try:
        elements = driver.find_elements(By.CLASS_NAME,'YmvwI')
        time.sleep(2)
        if num<len(elements):
            click_button(elements[num],driver)
        else : 
            print(f"탭을 찾을 수 없습니다: {num}")
    except Exception as e:
        print(f"Error occurred while clicking tab: {e}")

# 클래스 네임을 받아서 페이지에 모든 이미지 추출
def img_extration(class_name,driver,min_elements=100,folder_name = 'images'):
    try:
        # 정확히 일치하는 클래스 이름을 가진 이미지 요소 찾기
        img_elements = driver.find_elements(By.XPATH, f"//img[@class='{class_name}']")
        # 페이지를 스크롤하며 더 많은 요소 로드
        while len(img_elements) < min_elements:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # 페이지 로드 시간을 기다림
            img_elements = driver.find_elements(By.XPATH, f"//img[@class='{class_name}']")
        if not os.path.exists(folder_name):
            os.makedirs(folder_name)

        # 이미지의 src 속성 가져오기
        for index, img_element in enumerate(img_elements):
            # 이미지의 src 속성 가져오기
            img_src = img_element.get_attribute('src')

            if img_src.startswith('data:image/jpeg;base64,'):
                # base64로 인코딩된 이미지 처리
                img_data = img_src.split(',')[1]
                img_data = base64.b64decode(img_data)
                file_name = os.path.join(folder_name, f'image_{index + 1}.jpg')
                with open(file_name, 'wb') as f:
                    f.write(img_data)
                    
                print(f"이미지가 '{file_name}' 파일로 저장되었습니다.")
            elif img_src.startswith('http'):
                # URL로 된 이미지 처리
                img_data = requests.get(img_src).content
                file_name = os.path.join(folder_name, f'image_{index + 1}.jpg')
                with open(file_name, 'wb') as f:
                    f.write(img_data)
                    
                print(f"이미지가 '{file_name}' 파일로 저장되었습니다.")
            else:
                print(f"지원하지 않는 이미지 형식입니다: {img_src}")
    except Exception as e:
        print(f"Error occurred while extracting images: {e}")

# 메인 함수
def main():
    query = get_search_query()
    driver = google_search(query)
    cilck_tab(1,driver)
    img_extration("YQ4gaf",driver,500)
    print('프로그램 끝')
```

```python
# 프로그램 실행
if __name__ == "__main__":
    main()
```
